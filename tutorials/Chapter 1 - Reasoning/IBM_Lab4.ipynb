```python
from lnn import Propositions, And, Fact

# Rules
A, B, C, D, E = Propositions("A", "B", "C", "D", "E")
AND = And(A, B, C, D, E)


# Data
A.add_data(Fact.TRUE)
B.add_data(Fact.FALSE)
C.add_data(Fact.TRUE)
D.add_data(Fact.FALSE)
E.add_data(Fact.TRUE)
AND.add_data(Fact.TRUE)
```

```python
from lnn import Model, Loss, Direction

# Knowledge
model = Model()
model.add_knowledge(AND)

# Reasoning
model.infer()
AND.print()
```

```python
# Training
model.train(direction=Direction.UPWARD, losses=[Loss.CONTRADICTION])
AND.print(params=True)
```

```python
from lnn import Model, Proposition, And, Iff, Fact, Implies, Or
#Rules
A, B, C, D, E = Propositions("A", "B", "C", "D", "E")
IMPLIES = Implies(A,B)
CD = And(C,D)
IFFE = Iff(CD, E)
ROOT = And(IMPLIES, IFFE)


# Knowledge
model = Model()
model.add_knowledge(ROOT)



# Data
model.add_data({
    A: (0.2, 1.0),
    C: (1.0, 1.0),
    E: (0.3, 0.7)
})
# Reasoning
model.infer()
```
